```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "cat.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(8, 4))
plt.imshow(image, cmap='gray')
plt.axis("off")
plt.title("Gray")
plt.show()
```

```python
height = image.shape[0] // 3
width = image.shape[1]

B = image[:height, :]
G = image[height:2*height, :]
R = image[2*height:3*height, :]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(B, cmap='Blues')
axes[0].set_title("B")
axes[0].axis("off")

axes[1].imshow(G, cmap='Greens')
axes[1].set_title("G")
axes[1].axis("off")

axes[2].imshow(R, cmap='Reds')
axes[2].set_title("R")
axes[2].axis("off")

plt.show()
```

```python
RB_merged = np.dstack((R, np.zeros_like(R), B))
GB_merged = np.dstack((np.zeros_like(G), G, B))
RG_merged = np.dstack((R, G, np.zeros_like(R)))

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(RB_merged)
axes[0].set_title("R + B")
axes[0].axis("off")

axes[1].imshow(GB_merged)
axes[1].set_title("G + B")
axes[1].axis("off")

axes[2].imshow(RG_merged)
axes[2].set_title("R + G")
axes[2].axis("off")

plt.show()
```

```python
original_image = np.dstack((R, G, B))

plt.figure(figsize=(8, 4))
plt.imshow(original_image)
plt.axis("off")
plt.title("R + G + B")
plt.show()
```
