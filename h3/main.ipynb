```python
b,g,r = cv2.split(img_cat)

fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].imshow(r, cmap='Reds')
axes[0].set_title("Red Channel")
axes[1].imshow(g, cmap='Greens')
axes[1].set_title("Green Channel")
axes[2].imshow(b, cmap='Blues')
axes[2].set_title("Blue Channel")

for ax in axes:
    ax.axis("off")

plt.show()
```

```python
b,g,r = cv2.split(img_dog)

fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].imshow(r, cmap='Reds')
axes[0].set_title("Red Channel")
axes[1].imshow(g, cmap='Greens')
axes[1].set_title("Green Channel")
axes[2].imshow(b, cmap='Blues')
axes[2].set_title("Blue Channel")

for ax in axes:
    ax.axis("off")

plt.show()
```

```python
def arithmetic_operations(img, title):
    added = cv2.add(img, img)
    subtracted = cv2.subtract(img, img)
    multiplied = cv2.multiply(img, img)
    divided = cv2.divide(img,img)  
    
    plt.figure(figsize=(10, 8))
    
    plt.subplot(221), plt.imshow(cv2.cvtColor(added, cv2.COLOR_BGR2RGB)), plt.title(f'{title} - Addition')
    plt.axis("off")
    plt.subplot(222), plt.imshow(cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB)), plt.title(f'{title} - Subtraction')
    plt.axis("off")
    plt.subplot(223), plt.imshow(cv2.cvtColor(multiplied, cv2.COLOR_BGR2RGB)), plt.title(f'{title} - Multiplication')
    plt.axis("off")
    plt.subplot(224), plt.imshow(cv2.cvtColor(divided, cv2.COLOR_BGR2RGB)), plt.title(f'{title} - Division')
    plt.axis("off")
    
    plt.show()
```

```python
arithmetic_operations(img_dog, "Dog")
```

```python
arithmetic_operations(img_cat, "Cat")
```

```python

gray_img_cat = cv2.imread("cat.jpg", cv2.IMREAD_GRAYSCALE)

hist = cv2.calcHist([gray_img_cat], [0], None, [256], [0, 256])

plt.plot(hist, color="black")
plt.title("cat.png Grayscale Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
